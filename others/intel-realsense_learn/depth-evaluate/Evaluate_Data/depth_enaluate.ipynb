```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1,10,10)
y = np.linspace(1,10,10)

fig = plt.figure(figsize=(15, 10))

fig = plt.figure(1)

columns = 3
rows = 5

for i in range(1, columns*rows+1):
    fig.add_subplot(rows, columns, i)
    plt.scatter(x, y)

plt.tight_layout()
plt.show()
```
